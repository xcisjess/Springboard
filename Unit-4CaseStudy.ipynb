**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Calculate the mean by year of each borough, group by 'London_Borough' and 'Year', assign to a new dataframe dfg
dfg = df.groupby(['London_Borough', 'Year'])['Average_Price'].mean()
dfg.head()
```

```python
# Check the index, dfg has multi-level index
dfg = pd.DataFrame(dfg)
dfg.info()
dfg.index
```

```python
#Create a dictionary for ratios
final = {}
```

```python
#Use the list created before: Borough_list, to loop over the dataframe. Calculate the ratio for each borough.
for b in borough_list:
    ratio = dfg.loc[(b,1998)]/dfg.loc[(b,2018)]
    final[b] = ratio
print(final)
```

```python
#Make the dictionary final into a dataframe
df_ratio = pd.DataFrame(final)
df_ratio.head()
```

```python
#Transpose the df_ratio because the borough name is in columns
df_ratio_t = df_ratio.transpose()
df_ratio_t.head()
```

```python
#reset the index and add column headers
df_ratio_t = df_ratio_t.reset_index()
df_ratio_t.head()
```

```python
df_ratio_t.rename(columns={'index':'Borough', 'Average_Price':'2018'},inplace=True)
df_ratio_t.head()
```

```python
df_order = df_ratio_t.sort_values(by='2018',ascending=False)
df_order.head(32)
```

```python
#Add percentage to the ratio and reset the index
df_order['Percentage'] = round(df_order['2018']*100, 2)
df_order = df_order.reset_index()
df_order.head()
```

```python
#Drop the index
df_order = df_order.drop(df_order.columns[0], axis=1)
df_order.head()
```

```python
#Visualize the ratio
plot = df_order[['Borough','Percentage']].plot(kind='bar')

plot.set_xticklabels(df_order.Borough)
plot.set_ylabel('%')
```
